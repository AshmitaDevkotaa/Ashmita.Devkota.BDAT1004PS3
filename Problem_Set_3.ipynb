```python
# import the dataset
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
users = pd.read_csv(url, sep='|', index_col = 'user_id')
```

### Discover what is the mean age per occupation

```python
#  Discover the Male ratio per occupation and sort it from the most to the least

male_ratio_by_occupation = (users[users['gender'] == 'M']
                            .groupby('occupation')['gender']
                            .count() / users.groupby('occupation')['gender'].count()
                            ).sort_values(ascending=False)

male_ratio_by_occupation
```

####  Discover the Male ratio per occupation and sort it from the most to the least


```python
# Discover what is the mean age per occupation

mean_age_per_occupation = users.groupby('occupation')['age'].mean()
print("mean Age per Occupation",mean_age_per_occupation)
```

```python
#For each occupation, calculate the minimum and maximum ages
min_max_age_occ = users.groupby('occupation')['age'].agg(['min', 'max'])
min_max_age_occ
```

```python
# For each combination of occupation and sex, calculate the mean age
mean_age_by_occ_sex = users.groupby(['occupation','gender'])['age'].mean()
mean_age_by_occ_sex
```

```python
# For each occupation present the percentage of women and men

# Calculate the percentage of women and men for each occupation
gender_percentage_by_occupation = pd.crosstab(index=users['occupation'], columns=users['gender'], normalize='index') * 100
gender_percentage_by_occupation
```

```python
data = pd.read_table("wind.txt", sep='\s+', parse_dates=[[0,1,2]] )
data
```

```python
# Assign it to a variable called data and replace the first 3 columns by a proper datetime index
data.set_index('Yr_Mo_Dy', inplace=True)
```

```python
# Create a function to fix the year and apply it
def fix_year(x):
    if x.year > 2000:
        year = x.year - 100
    else:
        year = x.year
    return pd.to_datetime(f"{year}-{x.month}-{x.day}")
data.index = data.index.to_series().apply(fix_year)
```

```python
#Set the right dates as the index
data.index = pd.to_datetime(data.index)
data
```

```python
# Compute how many values are missing for each location over the entire record.They should be ignored in all calculations below
missing_values_per_location = data.isnull().sum()
```

```python
# Compute how many non-missing values there are in total.
total_non_missing_values = data.notnull().sum().sum()
```

```python
# Calculate the mean windspeeds of the windspeeds over all the locations and all the times
mean_windspeed = data.stack().mean()
mean_windspeed
```

```python
# Create a DataFrame called loc_stats and calculate the min, max and mean windspeeds and standard deviations of the windspeeds at each location over all the days
loc_stats = data.describe(percentiles=[]).transpose()
loc_stats
```

```python
# Create a DataFrame called day_stats and calculate the min, max and mean windspeed and standard deviations of the windspeeds across all the locations at each day.
#A different set of numbers for each day.
day_stats = data.resample('D').agg(['min', 'max', 'mean', 'std'])
day_stats
```

```python
# Find the average windspeed in January for each location. Treat January 1961 and January 1962 both as January.

# Extract the January data
january_data = data[data.index.month == 1]

# Group by both year and location and calculate the mean
january_average_per_location = january_data.groupby([january_data.index.year, january_data.index.month, january_data.index.day]).mean()
```

```python
#Downsample the record to a yearly frequency for each location
yearly_data = data.resample('Y').mean()
yearly_data
```

```python
# Downsample the record to a monthly frequency for each location
monthly_data = data.resample('M').mean()
monthly_data
```

```python
# Downsample the record to a weekly frequency for each location
weekly_data = data.resample('W').mean()
weekly_data
```

```python
# Calculate the min, max, mean, and standard deviations of the windspeeds across all locations for each week for the first 52 weeks

weekly_stats = data.resample('W').agg(['min', 'max', 'mean', 'std']).iloc[:52]
weekly_stats
```

#### Create a horizontal bar chart that compares the deadliest actors in Hollywood. Sort the actors by their kill count and label each bar with the corresponding actor's name. Don't forget to label your axes!

```python
#import necessary libraries 
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
#import emoji

#name the data frame impport from the csv file
actors_df = pd.read_csv("actor_kill_counts.csv")
```

```python
actors_df.head()
```

```python
#using sort values function to sort by count of the kills
sorted_df = actors_df.sort_values(by="Count")
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10,8))

#plots the bar using seaborn
ax= sns.barplot(x="Count",y="Actor",data=sorted_df)

#for labels on each bars
for i, count in enumerate(sorted_df['Count']):
    
    ax.text(count +8,i, str(count)+" Kills")
    
    
plt.xlabel("kill count")
plt.ylabel("Actor")
plt.title("Dangerous Hollywood Actors")
plt.tight_layout()
plt.show()
```

# Question 9

#### Create a pie chart showing the fraction of all Roman Emperors that were assassinated.
#### Make sure that the pie chart is an even circle, labels the categories, and shows the percentage breakdown of the categories.

```python
import pandas as pd
from matplotlib import pyplot as plt
#import seaborn as sns

emperors_df= pd.read_csv("roman-emperor-reigns.csv")
```

```python
emperors_df.head(11)
```

```python
#count the occurences of assasinated and non assasinated emperors

assassinate_emp = emperors_df[emperors_df["Cause_of_Death"]=="Assassinated"].count()["Cause_of_Death"]
non_assassinated_emp = emperors_df[emperors_df["Cause_of_Death"] != "Assassinated"].count()["Cause_of_Death"]

total_emperors = assassinate_emp+non_assassinated_emp

#percentage calculations

assasinated_emp_per = (assassinate_emp/total_emperors)*100
non_assassinated_emp_per = (non_assassinated_emp/total_emperors)*100

#data for pie chart
labels = "Assassinated Emperors", "Non-Assassinated emperors"
size = [assasinated_emp_per,non_assassinated_emp_per]
colors = ["red","green"]


#plotting chart
plt.figure(figsize=(8,8))
plt.pie(size,labels=labels, colors=colors, autopct="%1.2f%%", startangle=120)
plt.axis("equal")
plt.title("Fraction of Roman Emperors Assassinated")
plt.show()
```
